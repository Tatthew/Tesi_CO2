# Studio delle emissioni di CO2

Creiamo un dataframe contenente solo i dati relativi l'Italia

```python
it_df = df1[df1['Country'] == 'Italy']
it_df
```

Creiamo un dataframe che comprende solo i dati dal 1920 in poi

```python
it_CO2 = it_df[it_df['Year'] > 1920]
it_CO2
```

Grafico delle emissioni di CO2 in Italia dal 1920 al 2020

```python
sns.lineplot(data=it_CO2, x='Year', y='CO2 emission (Tons)').set(title='Italy CO2 emissions')
plt.show()
```

```python
fr_df = df1[df1['Country'] == 'France']
de_df = df1[df1['Country'] == 'Germany']
es_df = df1[df1['Country'] == 'Spain']
gr_df = df1[df1['Country'] == 'Greece']
pt_df = df1[df1['Country'] == 'Portugal']
gb_df = df1[df1['Country'] == 'United Kingdom']
nl_df = df1[df1['Country'] == 'Netherlands']
hu_df = df1[df1['Country'] == 'Hungary']
tr_df = df1[df1['Country'] == 'Turkey']
be_df = df1[df1['Country'] == 'Belgium']
dk_df = df1[df1['Country'] == 'Denmark']
fi_df = df1[df1['Country'] == 'Finland']
se_df = df1[df1['Country'] == 'Sweden']
```

Uniamo i dataframe dei paesi interessati in un unico dataframe

```python
dfs = [it_df, fr_df, de_df, es_df, gr_df, pt_df, gb_df, nl_df, hu_df, tr_df, be_df, dk_df, fi_df, se_df]
eu_df = pd.concat(dfs)
eu_df
```

Selezioniamo solo i dati degli ultimi 100 anni


```python
eu_CO2 = eu_df[eu_df['Year'] > 1920]
eu_CO2
```

```python
sns.lineplot(data=eu_CO2, x='Year', y='CO2 emission (Tons)', hue='Country').set(title='CO2 emissions from 1920 to 2020')
plt.show()
```

```python
eu_CO2_2020 = eu_CO2[eu_CO2['Year'] == 2020]
eu_CO2_2020 = eu_CO2_2020.replace('/km²', '', regex=True)
eu_CO2_2020['Density(km2)'] = eu_CO2_2020['Density(km2)'].astype('int')
eu_CO2_2020.sort_values(by='Density(km2)')
```

```python
sns.scatterplot(data=eu_CO2_2020, x='Density(km2)', y='CO2 emission (Tons)', hue='Country').set(title='CO2 emissions in relation to density')
plt.show()
```

```python
sns.scatterplot(data=eu_CO2_2020, x='Population(2022)', y='CO2 emission (Tons)', hue='Country').set(title='CO2 emissions in relation to population')
plt.show()
```

```python
sns.scatterplot(data=eu_CO2_2020, x='Area', y='CO2 emission (Tons)', hue='Country').set(title="CO2 emission in relation to the Country's Area")
plt.show()
```

```python
eu_CO2_2020['CO2 emission (Tons) per person'] = eu_CO2_2020['CO2 emission (Tons)'] / eu_CO2_2020['Population(2022)']
eu_CO2_2020['CO2 emission (Tons) per person'] = eu_CO2_2020['CO2 emission (Tons) per person'].astype('int')
eu_CO2_2020
```

```python
sns.barplot(data=eu_CO2_2020, x='Country', y='CO2 emission (Tons) per person', color='lightgreen').set(title='Tons of CO2 per person')
plt.xticks(rotation=45)
plt.show()
```
